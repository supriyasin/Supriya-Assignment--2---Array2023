```python
#<aside>
ðŸ’¡ **Question 1**
Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2),..., (an, bn) such that the 
 sum of min(ai, bi) for all i is maximized. Return the maximized sum.

**Example 1:**
Input: nums = [1,4,3,2]
Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3
2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3
3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4
So the maximum possible sum is 4
</aside>

"""To solve this problem, we can sort the given array nums in ascending order. This way, the minimum elements will be paired 
   with each other, and the maximum elements will be paired with each other, which helps maximize the sum of the minimum values
   in each pair.

   Here's the step-by-step algorithm:
   
   1. Sort the array nums in ascending order.
   
   2. Initialize a variable max_sum to 0 to keep track of the maximum sum.
   
   3. Iterate over the array nums by incrementing the index by 2 each time (since we need to form pairs).
   
   4. For each pair of numbers (a, b) at indices i and i+1, add the minimum value between a and b to max_sum.
   
   5. Return the value of max_sum."""

#Here's the implementation in Python:

def array_pair_sum(nums):
    nums.sort()
    max_sum = 0
    for i in range(0, len(nums), 2):
        max_sum += min(nums[i], nums[i+1])
    return max_sum

#Let's test the function with the given example:

nums = [1, 4, 3, 2]
print(array_pair_sum(nums)) #The maximum possible sum of the minimum values in pairs is 4, which matches the expected output.
```

```python
#Question 2
Alice has n candies, where the ith candy is of type candyType[i]. Alice noticed that she started to gain weight, so she 
visited a doctor. 

The doctor advised Alice to only eat n / 2 of the candies she has (n is always even). Alice likes her candies very much,\
and she wants to eat the maximum number of different types of candies while still following the doctor's advice. 

Given the integer array candyType of length n, return the maximum number of different types of candies she can eat if she 
only eats n / 2 of them.

Example 1:
Input: candyType = [1,1,2,2,3,3]
Output: 3

Explanation: Alice can only eat 6 / 2 = 3 candies. Since there are only 3 types, she can eat one of each type.
    
"""To solve this problem, we can use a set to keep track of the different types of candies Alice has. We iterate over the 
   candyType array and add each candy type to the set. Since a set only stores unique elements, at the end, the set will
   contain all the different types of candies Alice has.

   To calculate the maximum number of different types of candies Alice can eat, we compare the size of the set with half
   the total number of candies (n/2). If the set size is smaller than or equal to n/2, Alice can eat all the different 
   types of candies she has. Otherwise, she can only eat n/2 different types of candies."""

#Here's the implementation in Python:


def distributeCandies(candyType):
    max_types = len(set(candyType))
    max_candies = len(candyType) // 2
#Using this implementation, for the given example, distributeCandies([1, 1, 2, 2, 3, 3]) would return 3, indicating that
 Alice can eat three different types of candies.
    
    return min(max_types, max_candies)
candyType = [1, 1, 2, 2, 3, 3]
result = distributeCandies(candyType)
print(result) #When you run this code, it will output 3, indicating that Alice can eat three different types of candies from 
              the given candyType list. 
```

```python
#Question 3
We define a harmonious array as an array where the difference between its maximum value
and its minimum value is exactly 1.

Given an integer array nums, return the length of its longest harmonious subsequence
among all its possible subsequences.

A subsequence of an array is a sequence that can be derived from the array by deleting some or no elements without 
changing the order of the remaining elements.

Example 1:
Input: nums = [1,3,2,2,5,2,3,7]
Output: 5

Explanation: The longest harmonious subsequence is [3,2,2,2,3].

from collections import Counter

"""To find the length of the longest harmonious subsequence in the given array, you can follow the following steps:

    1. Initialize a dictionary 'freq' to count the frequency of each number in the array.
    
    2. Iterate through each number 'num' in the array and update its frequency in the 'freq' dictionary.
    
    3. Initialize a variable 'max_length' to 0.
    
    4. Iterate through each number 'num' in the array again.
         . If 'num + 1' exists in the 'freq' dictionary, calculate the length of the harmonious subsequence starting with 
           'num' by adding the frequencies of 'num' and 'num + 1'.
         . If the length is greater than the current 'max_length', update 'max_length'.
         
    5. Return 'max_length'."""

#Here's the implementation of the above algorithm in Python:

def findLHS(nums):
    freq = Counter(nums)
    max_length = 0

    for num in nums:
        if num + 1 in freq:
            length = freq[num] + freq[num + 1]
            if length > max_length:
                max_length = length

#Using the given example:

    return max_length
nums = [1, 3, 2, 2, 5, 2, 3, 7]
result = findLHS(nums)
print(result)  # Output: 5
```

```python
#Question 4
You have a long flowerbed in which some of the plots are planted, and some are not.
However, flowers cannot be planted in adjacent plots.
Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty, and an integer n, 
return true if n new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false 
otherwise.

Example 1:
Input: flowerbed = [1,0,0,0,1], n = 1
Output: true
    
"""To solve this problem, we can iterate through the flowerbed array and check if each plot is empty (0), as well as the 
adjacent plots. If a plot is empty and its adjacent plots are also empty or out of bounds, we can plant a flower in that plot. 
We repeat this process until we have planted the required number of flowers or we have traversed the entire flowerbed."""

#Here's the implementation in Python:

def can_place_flowers(flowerbed, n):
    count = 0
    i = 0
    while i < len(flowerbed):
        if flowerbed[i] == 0 and (i == 0 or flowerbed[i - 1] == 0) and (i == len(flowerbed) - 1 or flowerbed[i + 1] == 0):
            flowerbed[i] = 1
            count += 1
        i += 1
        if count >= n:
            return True
#Let's test the provided example:

return False
flowerbed = [1, 0, 0, 0, 1]
n = 1
print(can_place_flowers(flowerbed, n))

#In this case, we can place 1 new flower in the empty plot at index 1, and the resulting flowerbed would be [1, 1, 0, 0, 1], 
 which satisfies the condition of no adjacent flowers. Therefore, the function returns True.
```

```python
#Question 5
Given an integer array nums, find three numbers whose product is maximum and return the maximum product.

Example 1:
Input: nums = [1,2,3]
Output: 6
    
"""To find the maximum product of three numbers in an integer array, you can follow the following approach:

     1. Sort the input array in non-decreasing order.
     
     2. Calculate the product of the last three elements in the sorted array (since they will be the largest numbers).
     
     3. Calculate the product of the first two elements (the smallest) and the last element (the largest) in the sorted array.
     
     4. Compare the two products and return the maximum."""

#Here's the Python code that implements this approach:

def maximumProduct(nums):
    nums.sort()
    n = len(nums)
    return max(nums[n-1] * nums[n-2] * nums[n-3], nums[0] * nums[1] * nums[n-1])

#Using this code, for the example input nums = [1, 2, 3], the maximum product would be calculated as follows:

 1. Sort the array: [1, 2, 3]
 2. Calculate the product of the last three elements: 2 * 3 * 1 = 6
 3. Calculate the product of the first two elements and the last element: 1 * 2 * 3 = 6
 4. Return the maximum of the two products: max(6, 6) = 6, which is the expected output.
    
nums = [1, 2, 3]
result = maximumProduct(nums)
print(result)
```

```python
#Question 6
Given an array of integers nums which is sorted in ascending order, and an integer target,
write a function to search target in nums. If target exists, then return its index. Otherwise,
return -1.

You must write an algorithm with O(log n) runtime complexity.

Input: nums = [-1,0,3,5,9,12], target = 9
Output: 4

Explanation: 9 exists in nums and its index is 4
    
"""To solve this problem with a runtime complexity of O(log n), we can use the binary search algorithm. Here's an 
   implementation in Python:"""  

"""Let's go through the code:

     1. We initialize two pointers, left and right, to the first and last indices of the nums array, respectively.
     
     2. We enter a while loop that continues as long as left is less than or equal to right. This condition ensures 
        that the search space is not empty.
        
     3. Inside the loop, we calculate the middle index mid using integer division. This will give us the index of the 
        element in the middle of the current search space.
        
     4. We compare the element at the mid index with the target. If they are equal, we have found the target, so we return 
        the index mid.
        
     5. If the element at mid is less than the target, we update left to mid + 1 to search the right half of the current 
        search space. 
        
     6. If the element at mid is greater than the target, we update right to mid - 1 to search the left half of the current 
        search space.
        
     7. If we exit the loop without finding the target, we return -1 to indicate that the target does not exist in the array.
     
     8. This algorithm divides the search space in half at each iteration, resulting in a logarithmic time complexity of 
        O(log n), where n is the size of the input array nums."""

def search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
nums = [-1, 0, 3, 5, 9, 12]
target = 9

result = search(nums, target)
print(result)
```

```python
#To determine whether an array is monotonic, we can iterate through the array andcheck if it satisfies either the 
 monotone increasing or monotone decreasing conditions. Here's the Python code to solve this:
    
"""In this code, we initialize two boolean variables, increasing and decreasing, as True to track whether the array is 
   monotone increasing or monotone decreasing. We iterate through the array starting from the second element and compare 
   it with the previous element.

   If we find an element that is less than the previous element, we set increasing to False, as it violates the monotone 
   increasing condition. Similarly, if we find an element that is greater than the previous element, we set decreasing to 
   False, as it violates the monotone decreasing condition.

   Finally, we return increasing or decreasing, which will be True if the array is monotone (either increasing or decreasing) 
   and False otherwise.

   For the given example [1, 2, 2, 3], the output will be True since the array is monotone increasing."""    

def isMonotonic(nums):
    increasing = decreasing = True

    for i in range(1, len(nums)):
        if nums[i] < nums[i - 1]:
            increasing = False
        if nums[i] > nums[i - 1]:
            decreasing = False

    return increasing or decreasing

# Test case
nums = [1, 2, 2, 3]
print(isMonotonic(nums))  # Output: True
```

```python
#Question 8
You are given an integer array nums and an integer k.

In one operation, you can choose any index i where 0 <= i < nums.length and change nums[i] to nums[i] + x where x is an 
integer from the range [-k, k]. You can apply this operation at most once for each index i.

The score of nums is the difference between the maximum and minimum elements in nums.

Return the minimum score of nums after applying the mentioned operation at most once for each index in it.

Example 1:
Input: nums = [1], k = 0
Output: 0

Explanation: The score is max(nums) - min(nums) = 1 - 1 = 0.

"""To find the minimum score of the array nums after applying the given operation, we can follow these steps:

     1. Find the minimum and maximum values in the original array nums.
     
     2. Calculate the difference between the maximum and minimum values, which represents the initial score.
     
     3. If k is 0, return the initial score since we cannot make any changes.
     
     4. Calculate the potential minimum and maximum values after applying the operation. The potential minimum would be 
        the minimum value minus k, and the potential maximum would be the maximum value plus k.
        
     5. Update the minimum score by comparing the initial score with the difference between the potential maximum and 
         potential minimum.
         
     6. Return the updated minimum score."""

#Here's the implementation of the solution in Python:
    

def minimumScore(nums, k):
    minimum = min(nums)
    maximum = max(nums)
    initial_score = maximum - minimum

    if k == 0:
        return initial_score

    potential_minimum = minimum - k
    potential_maximum = maximum + k
    updated_score = potential_maximum - potential_minimum

    return min(initial_score, updated_score)

#Let's test the example provided:

nums = [1]
k = 0
print(minimumScore(nums, k))

#As expected, the output is 0, which is the minimum score of the array nums after applying the operation
(which, in this case, does not change the array).
```
